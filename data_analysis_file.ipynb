**Data preparation Questions**

1. Checking if there are any null values in the dataset and filling it if there are null values

```python
dataset.isnull().sum()
```

2. Checking if there are any duplicates in the dataset

```python
duplicates = dataset.duplicated(["Product ID", "Product Name"])
print(f"Number of duplicates: {duplicates.sum()}")
dataset = dataset.drop_duplicates(["Product ID", "Product Name"])
```

```python
total_rows = len(dataset)
unique_values = dataset['Product ID'].nunique() # Select the 'Product ID' column for unique value count

print(f"Total rows: {total_rows}")
print(f"Unique values in 'ID': {unique_values}")

if unique_values < total_rows:
    print("'ID' column contains duplicates.")
else:
    print("'ID' column has no duplicates.")
```

3. Ensuring the datatype of columns is accurate.

```python
print("Datatype of 'Age' column:", dataset['Age'].dtype)
print("Datatype of 'Price' column:", dataset['Price'].dtype)
print("Datatype of 'Popularity' column:", dataset['Popularity'].dtype)
print("Datatype of 'Rating' column:", dataset['Rating'].dtype)
```

4. Checking the inconsistencies

```python
dataset.describe()
```

```python
dataset.value_counts().sum()
```

```python
unique_brand_types = dataset["Brand Name"].unique()
print("Unique Brand Names:", unique_brand_types)
```

```python
dataset["Brand Name"] = dataset["Brand Name"].str.strip().str.title()
dataset["Country of Origin"] = dataset["Country of Origin"].str.strip().str.upper()
```

```python
brand_mapping = {"gucci": "Gucci", "GUCCI": "Gucci"}
country_mapping = {"USA": "United States", "U.S.A.": "United States"}
dataset["Brand Name"] = dataset["Brand Name"].replace(brand_mapping)
dataset["Country of Origin"] = dataset["Country of Origin"].replace(country_mapping)
```

```python
print("Standardized Brand Types:", dataset["Brand Name"].unique())
print("Standardized Country Types:", dataset["Country of Origin"].unique())
```

5. Clothing type consistency

```python
unique_clothing_types = dataset["Clothing Type"].unique()
print("Unique Clothing Types:", unique_clothing_types)
```

```python
# optional: only if variation existed
dataset["Clothing Type"] = dataset["Clothing Type"].str.strip().str.title()
mapping = {"jeans": "Jeans", "tshirt": "T-Shirt", "blazer": "Blazer"}
dataset["Clothing Type"] = dataset["Clothing Type"].replace(mapping)
```

```python
print("Standardized Clothing Types:", dataset["Clothing Type"].unique())
```

```python
import matplotlib.pyplot as plt

material_gender_price = dataset.groupby(['Clothing Type','Gender'])['Age'].mean().unstack()

plt.figure(figsize=(12, 6))
for gender in material_gender_price.columns:
    plt.plot(
        material_gender_price.index,
        material_gender_price[gender],
        marker='o',
        label=gender
    )

plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

material_gender_price = dataset.groupby(['Material', 'Gender'])['Price'].mean().unstack()

plt.figure(figsize=(12, 6))
for gender in material_gender_price.columns:
    plt.plot(
        material_gender_price.index,
        material_gender_price[gender],
        marker='o',
        label=gender
    )

plt.title('Average Price by Material and Gender', fontsize=14)
plt.xlabel('Material', fontsize=12)
plt.ylabel('Average Price', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Gender', loc='upper right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
